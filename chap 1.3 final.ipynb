```python
import cv2
import numpy as np

drawing = False
mode = True 
ix,iy = -1,-1
font = cv2.FONT_HERSHEY_SIMPLEX
cx,cy=-1,-1
# mouse callback function
def draw_circle(event,x,y,flags,param):
    
    global ix,iy,cx,cy,drawing,mode
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
        cx,cy=x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),0)
                cx,cy=x,y
 
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),0)
        cx,cy=x,y
def nothing(x):
    pass

#img = np.zeros((512,512,3), np.uint8)
img = cv2.imread('test2.jpg')
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)
cv2.createTrackbar('value', 'image',0,255,nothing)

while(1):
    img = cv2.imread('test2.jpg')
    
    k = cv2.waitKey(1) & 0xFF
    if k == ord('m'):
        mode+=1
        if mode>3:
            mode=0
    elif k == 27:
        break
    value = cv2.getTrackbarPos('value','image')
    if ix!=-1 and iy!=-1:
        cv2.rectangle(img,(ix,iy),(cx,cy),(0,255,0),0)
        if mode ==1:
            clip=img[iy:cy,ix:cx]
            clip[:,:,1]=0
            clip[:,:,2]=0
        if mode ==2:
            clip=img[iy:cy,ix:cx]
            clip[:,:,0]=0
            clip[:,:,2]=0
        if mode ==3:
            clip=img[iy:cy,ix:cx]
            clip[:,:,0]=0
            clip[:,:,1]=0
    msg="Mouse-("+str(ix)+","+str(iy)+")-("+str(cx)+","+str(cy)+")"
    cv2.putText(img,msg,(10,500), font, 1,(255,255,255),2,cv2.LINE_AA)
    cv2.imshow('image',img)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
BLUE = [255,0,0]
img1 = cv2.imread('test4.jpg')
m=300
img1.shape
```

```python
replicate = cv2.copyMakeBorder(img1,m,m,m,m,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,m,m,m,m,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,m,m,m,m,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,m,m,m,m,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,m,m,m,m,cv2.BORDER_CONSTANT,value=BLUE)
plt.figure(figsize=(9,6))
plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```
