```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img1 = cv2.imread('xim0.png',0) # queryImage
img2 = cv2.imread('xim1.png',0) # trainImage
```

```python
sift1 = cv2.SIFT_create()
sift2 = cv2.SIFT_create()
# find the keypoints and descriptors with SIFT
kp1, des1 = sift1.detectAndCompute(img1,None) ## 키포인트 찍기
kp2, des2 = sift2.detectAndCompute(img2,None)
```

```python
# BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2, k=2)  ## 그림 2개를 디스크립트 넣고 knn 방식으로
# 메치를 시킨다.
```

```python
# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.15*n.distance: ## 두 매치 사이 거리가 0.15 보다 작은것 을 뽑아넨다.
        good.append([m])
```

```python
print(good)
```

```python
# cv2.drawMatchesKnn expects list of lists as matches.
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=2)
#이미지 1 과 이미지 2  키포인트 들을 정한후 아까 찾은 값들을 줄 2로 출력 
```

```python
plt.figure(figsize=(30,30))
plt.subplot(121), plt.imshow(img3, 'gray')
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('son2.png',0) # queryImage
img2 = cv2.imread('xr2.png',0) # trainImage
```

```python
# Initiate SIFT detector
sift = cv2.SIFT_create()
```

```python
# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
```

```python
kp2, des2 = sift.detectAndCompute(img2,None)
```

```python
# BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2, k=2)
# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.85*n.distance:
        good.append([m])
```

```python
# cv2.drawMatchesKnn expects list of lists as matches.
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=2)
plt.imshow(img3),plt.show()
```
